```python
from IPython.display import clear_output
clear_output()

game_on = True
game_list = [0,1,2]

def display_game(game_list):
  print('Here is your current list')
  print(game_list)


def position_choice():
  choice = 'wrong'

  while choice not in ['0','1','2']:
    choice = input("Pick a position (0,1,2): ")

    if choice not in ['0','1','2']:
      clear_output()
      print("Sorry you have not enter value within range")

  return int(choice)

def replacement_choice(game_list,position):

  user_choice = input("Input your string: ")
  game_list[position] = user_choice

  return game_list

def gameon_choice():
  choice = 'wrong'

  while choice not in ['Y','N']:
      choice = input("Would you like to play again? Enter Y or N: ")

      if choice not in ['Y','N']:
        clear_output()
        print("Sorry I do not understand your choice, please enter Y or N")

  if choice =='Y':
      return True

  else:
      return False

while game_on:
  clear_output()
  display_game(game_list)
  position = position_choice()
  game_list = replacement_choice(game_list,position)
  clear_output()
  display_game(game_list)
  game_on = gameon_choice()
```
